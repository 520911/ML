Целевая переменная – тип стекла «Type». Остальные признаки описывают химические элементы в составе материала. Датасет нужно исследовать на наличие выбросов, провести EDA.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('glass.csv')
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
sns.countplot(df['Type'])
```

Видим, что классы несбалансированы.

```python
features = df.iloc[:, :-1]
features.head()
```

```python
target = df.iloc[:, -1:]
target.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, np.ravel(y_train))
y_pred = rf.predict(X_test)
rf.score(X_test, y_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=features)
```

```python
sns.pairplot(features)
# Замечаем выброы в некоторых признаках: Na, Al, Si, K, Ca, Ba
```

```python
df = pd.concat([features, target], axis=1)
df.shape
```

```python
# Найдем индексы строк с выбросами
outliers_indexes = []

for feature in df.iloc[:, :-1]:
  Q1 = np.percentile(df[feature], 15)
  Q3 = np.percentile(df[feature], 85)
  IQR = Q3 - Q1
  outliers = 1.5 * IQR
  outliers_indexes.append(df[(df[feature] < Q1 - outliers) | (df[feature] > Q3 + outliers)].index.values.tolist())
```

```python
outliers_indexes = list(set(sum(outliers_indexes, [])))
```

```python
df.shape
```

```python
df = df.drop(outliers_indexes).reset_index(drop=True)
```

```python
df.shape
# Видим, что из 214 строк осталось 173
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=df.iloc[:,:-1])
```

```python
features_new = df.iloc[:,:-1]
target_new = df.iloc[:,-1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_new, target_new, test_size=0.2, random_state=42)
```

```python
rfn = RandomForestClassifier()
rfn.fit(X_train, np.ravel(y_train))
y_pred = rfn.predict(X_test)
rfn.score(X_test, y_test)
# Метрика ухудшилась
```

```python
confusion_matrix(y_test, y_pred)
```
